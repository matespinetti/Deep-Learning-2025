```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('../data/practica3/dataset_bits_decimal.csv')
df = df_original.copy()

df.head()
```

```python
x = df.drop(columns=['decimal'])
y = df['decimal']

```

```python
from linearreg import LinearSGDRegressor

ml = LinearSGDRegressor(lr=0.05, epochs=1000, verbose=True, tol=1e-10)
ml.fit(x,y)
print(ml.get_params())
print(ml.score(x,y))
```

```python
ml = LinearSGDRegressor(lr=0.05, epochs=1000, verbose=True, tol=1e-10, fit_intercept=False)
ml.fit(x,y)
print(ml.get_params())
print(ml.score(x,y))
```

### Comparacion de modelos


Con intercept:
El modelo tiene un parametro extra(bias) que aprednder. Como el problema ya es perfectamente representable con pesos exactos, el bias no es escritctamente necesario. Por eso, el entanmiento necesita algunas iteraciones mas para ajustar tanto los pesos como el bias y "empujar" el bias hacia cero.


Sin intecept: El modelo solo entrena tres pesos. Como la solucion perfecta no requiere bias, converge mas rapido al vector correcto. Menos parametros = menos grados de libertad = menos pasos de ajuste
