```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd

# We will reuse X, y as defined above
# y is binary (0/1) here, as per your current notebook

lrs = [0.1, 0.2, 0.3]
train_ratios = [0.5, 0.6, 0.7, 0.8, 0.9]
combos = [
    ("sigmoid", "mse"),      # i) sigmoid + ECM
    ("sigmoid", "bce"),      # ii) sigmoid + EC_binaria
    ("tanh",    "mse"),      # iii) tanh + ECM
]

n_runs = 3

results = []

for activation, cost in combos:
    for train_ratio in train_ratios:
        for lr in lrs:
            train_accs = []
            test_accs = []
            iters = []
            for run in range(n_runs):
                X_train, X_test, y_train, y_test = train_test_split(
                    X, y, train_size=train_ratio, stratify=y, random_state=run
                )
                scaler = StandardScaler()
                X_train_sc = scaler.fit_transform(X_train)
                X_test_sc  = scaler.transform(X_test)

                model = MulticlassRN(learning_rate=lr, max_iter=400, activation=activation, cost=cost, verbose=False)
                model.fit(X_train_sc, y_train)

                train_acc = model.score(X_train_sc, y_train)  # proporción 0..1
                test_acc  = model.score(X_test_sc,  y_test)

                train_accs.append(train_acc)
                test_accs.append(test_acc)

                # iterations performed: prefer cost_history/errors_ if exists; else fallback to max_iter
                it = getattr(model, 'cost_history', None)
                if it is None:
                    it = getattr(model, 'errors_', None)
                iters.append(len(it) if it is not None else 400)

            results.append({
                'activation': activation,
                'cost': cost,
                'train_ratio': train_ratio,
                'learning_rate': lr,
                'train_acc_mean': float(np.mean(train_accs)),
                'test_acc_mean':  float(np.mean(test_accs)),    
                'iters_mean': float(np.mean(iters)),
                'iters_std':  float(np.std(iters, ddof=1)),
            })

summary_df = pd.DataFrame(results).sort_values(['activation','cost','train_ratio','learning_rate']).reset_index(drop=True)
summary_df
```
