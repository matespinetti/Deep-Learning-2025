No hay nulos y las clases target estan equilibradas. Seguimos con division en train_test

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Clase'])
y = (df['Clase'] == 'Tipo2').astype(int)
```

```python
from sklearn.preprocessing import StandardScaler


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


```

```python
X_train_scaled.shape, y_train.shape
```

```python
lr = 0.04
for i in range(20):
    ml = LogisticRegression(learning_rate=lr, max_iter=20, tolerance=1e-28)
    ml.fit(X_train_scaled, y_train)
    print("Learning rate utilizado: ", lr)
    print( "Accuracy del modelo:", ml.score(X_test_scaled, y_test))
    print("Pesos asignados y bias: ", ml.get_weights())
  
    ml.plot_cost_history()
    lr += 0.04
```

### Usando funcion tangente

```python
# y_train_tanh = 2 * y_train - 1 NO HACE FALTA LO HACE SOLO EL REGRESOR
# y_test_tanh = 2 * y_test - 1
lr = 0.04
for i in range(20):
    ml = LogisticRegression(learning_rate=lr, max_iter=20, tolerance=1e-28, activation='tanh')
    ml.fit(X_train_scaled, y_train)
    print("Learning rate utilizado: ", lr)
    print( "Accuracy del modelo:", ml.score(X_test_scaled, y_test))
    print("Pesos asignados y bias: ", ml.get_weights())
  
    ml.plot_cost_history()
    lr += 0.04

```
