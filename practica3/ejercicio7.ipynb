```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('../data/automobile-simple.csv')
df = df_original.copy()

df.head()
```

```python
df.describe(include='all')
```

```python
df['num-of-doors'] = df['num-of-doors'].map({'two': 2, "four": 4})
df.head()
```

```python
df.isnull().sum()
```

```python
df['num-of-doors'] = df['num-of-doors'].fillna(df['num-of-doors'].mode()[0])
df['horsepower'] = df['horsepower'].fillna(df['horsepower'].mean())
```

```python
df = df.dropna()
df.describe(include='all')
```

```python
numeric_df = df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
corr_matrix
```

Como podemos observar en la matriz de correlacion, highway-mpg y city-pmg estan muy fuertemente relacionadas(0.97), por lo cual vamos a quedarnos con solo una de ellas. Luego procedemos a codificar categorias y estandarizar

```python
df = df.drop(columns=['highway-mpg'])
```

### Estandarizacion de variables numericas

```python
from sklearn.preprocessing import StandardScaler


columns_to_estandarize = ['curb-weight', 'engine-size', 'horsepower', 'city-mpg', 'volume', 'eco-rating']
scaler = StandardScaler()
df[columns_to_estandarize] = scaler.fit_transform(df[columns_to_estandarize])

df.head()
```

```python
categorical_columns = ['make', 'fuel-type', 'body-style']
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)

df.dtypes
```

```python
from sklearn.model_selection import train_test_split
from linearreg import LinearSGDRegressor
X = df.drop(columns=['price'])
y = df['price']


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

ml = LinearSGDRegressor( lr=0.05, epochs=1000, verbose=True)
ml.fit(X_train, y_train)
print(ml.get_params())

print(ml.score(X_test, y_test))

```
