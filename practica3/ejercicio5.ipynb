```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_original = pd.read_csv('../data/automobile-simple.csv')
df = df_original.copy()
df
```

```python
print("Correlacion lineal entre engine size y price")
print(df['engine-size'].corr(df['price']))
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler


df['price'] = df['price'].fillna(df['price'].mean())
scaler = StandardScaler()
df['engine-size'] = scaler.fit_transform(df[['engine-size']])
```

```python
## Normalizando


from linearreg import LinearSGDRegressor


x = df['engine-size']
y = df['price']


ml = LinearSGDRegressor(lr=0.01, epochs=1000, verbose=True, shuffle=True, random_state=42)
ml.fit(x, y)


print(ml.get_params())
print(ml.score(x,y))

y_pred = ml.predict(x)



```

```python
plt.scatter(x,y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red')
plt.xlabel('engine-size(estandarizado)')
plt.ylabel('price')
plt.legend()
plt.show()
```
