## Ejercicio 3

### Diagrama de barras
- Comparación de frecuencias o valores entre diferentes categorías discretas
- Distribución de una variable categórica
- Permite identificar cuál categoría es más o menos frecuente
- Ideal para variables nominales u ordinales


```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.countplot(data=df, x='NObesity', palette='viridis')
plt.title('Distribución de Obesidad')
plt.xlabel('Nivel de Obesidad')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Histograma
Un histograma muestra la distribución de una variable numérica continua. Te permite ver:

- Dónde se concentran la mayoría de los valores
- Si la distribución es simétrica o sesgada
- Si hay valores atípicos
- La forma general de los datos

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title('Distribución de Edad')
plt.xlabel('Edad')
plt.ylabel('Cantidad')
plt.show()
```

### Diagrama de caja
Un boxplot muestra el resumen estadístico de una variable numérica:
- Mediana (línea central)
- Cuartiles Q1 y Q3 (bordes de la caja)
- Valores mínimo y máximo (bigotes)
- Valores atípicos (puntos fuera de los bigotes)



```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='Gender', x='Weight')
plt.title('Distribución del Peso por Género')
plt.xlabel('Género')
plt.ylabel('Peso')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Diagrama de dispersion
Un scatterplot muestra la relación entre dos variables numéricas:

- Si hay correlación (positiva, negativa o ninguna)
- Qué tan fuerte es la relación
- Si hay patrones o grupos en los datos
- Identificar valores atípicos

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Height', y='Weight', hue='Gender', palette='viridis')
plt.title('Relación entre Altura y Peso')
plt.xlabel('Altura')
plt.ylabel('Peso')
plt.tight_layout()
plt.show()
```

## Ejercicio 6

### Division por rango


```python
##Por rango
df['CH2O_rango'] = pd.cut(df['CH2O'], bins=3, labels=['bajo', 'medio', 'alto'])

print(df['CH2O_rango'].value_counts().sort_index())

intervalos = pd.cut(df['CH2O'], bins=3)
print("Intervalos creados:")
print(intervalos.cat.categories)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='CH2O_rango', palette='viridis')
plt.title('Distribución de CH2O por rango')
plt.show()



```

```python
## Por intervalos 
df['CH20_intervalo'] = pd.qcut(df['CH2O'], q=3, labels=['bajo', 'medio', 'alto'])
print('=== Distribución de CH20 por intervalos ===')
print(df['CH20_intervalo'].value_counts().sort_index())

print("Intervalos creados:")
print(pd.qcut(df['CH2O'], q=3).cat.categories)

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='CH20_intervalo', palette='viridis')
plt.title('Distribución de CH20 por intervalos')
plt.show()



```

```python

Q1 = df["Height"].quantile(0.25)
Q3 = df["Height"].quantile(0.75)
IQR = Q3 - Q1

lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR

# Contar personas con altura atípica
outliers = df[(df["Height"] < lim_inf) | (df["Height"] > lim_sup)]
cantidad_outliers = outliers.shape[0]

print("Cantidad de personas con altura atípica:", cantidad_outliers)
```
